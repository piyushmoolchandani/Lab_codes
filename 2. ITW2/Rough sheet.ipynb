```python
import math
a = input()
count = {}
for i in a:
    if i not in count:
        count[i] = 1
    else:
        count[i] += 1
p = math.ceil(sum(count.values())/ len(count))
ans = 0
for i in count.keys():
    ans += count[i] - p
print(abs(ans))
```
